# plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../data/cluster_data.csv')
```

```python
sample_df = df[df['reducer']=='None']
```

```python
# line plot of DBS score vs. sample size colored by vectorizer
sns.lineplot(x='size', y='DBS', hue='vectorizer', data=sample_df, markers=True)
# add dots to line plot
sns.scatterplot(x='size', y='DBS', hue='vectorizer', data=sample_df, s=100, alpha=0.5)

# plot title and labels
plt.title('DBS vs. Sample Size')
plt.xlabel('Sample Size')
plt.ylabel('DBS')
```

```python
vec_red_dbs = df[df['reducer']!='None']
```

```python
# plot DBS score vs. vectorizer and reducer
sns.barplot(x='vectorizer', y='DBS', hue='reducer', data=vec_red_dbs)

# plot title and labels
plt.title('DBS vs. Vectorizer and Reducer')
```
